```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import cm
from scipy import stats
from collections import deque
from scipy.optimize import minimize
from scipy.linalg import norm, inv, det
import time
import seaborn as sns
from multiprocess import Pool, cpu_count
import pickle 

sns.set_theme()
```

```python
import pickle
with open('estimationsSingleFullModel2410.pkl', 'rb') as fi:
    _, estimations_max, estimations = pickle.load(fi)
    
mu = np.array([[1.0],
               [1.0]])

alpha = np.array([[0.0, 0.0],
                  [0.5, 0.0]])

beta = np.array([[0.0],
                 [1.3]])

noise = 0.5

single_theta = np.concatenate((mu.ravel(), alpha.ravel(), beta.ravel(), np.array([noise])))
```

```python
parameters_full = [r"$\mu_1$", r"$\mu_2$", r"$\alpha_{11}$", r"$\alpha_{12}$", r"$\alpha_{21}$", r"$\alpha_{22}$", r"$\beta_1$", r"$\beta_2$", r"$\lambda_0$"]

fig, ax = plt.subplots(1, 1, figsize=(12,8))
_ = ax.boxplot(estimations)#, showmeans=True)
_ = ax.scatter(range(1, 10), single_theta, marker="*", s=100)
ax.set_ylim((-0.1,3))
ax.set_xticklabels(parameters_full)
print("Proportion of null:", np.mean(estimations < 2e-16, axis=0))
_=ax.set_title("Full model")
```

```python
parameters_reduced = [r"$\mu_1$", r"$\mu_2$", r"$\alpha_{21}$",r"$\beta_2$", r"$\lambda_0$"]

fig, ax = plt.subplots(1, 1, figsize=(12,8))
_ = ax.boxplot(estimations_max)#, showmeans=True)
aux_single_theta = single_theta[single_theta > 0.0]
print(aux_single_theta.shape)
_ = ax.scatter(range(1,6), aux_single_theta, marker="*", s=100)
ax.set_xticklabels(parameters_reduced)

print("Proportion of null:", np.mean(estimations_max < 2e-16, axis=0))
_ = ax.set_title("Reduced model")
```

```python
import pickle
with open('estimationsSingleNonNoisedModel2410.pkl', 'rb') as fi:
    _, estimations_nonnoised = pickle.load(fi)
```

```python
parameters_full = [r"$\mu_1$", r"$\mu_2$", r"$\alpha_{11}$", r"$\alpha_{12}$", r"$\alpha_{21}$", r"$\alpha_{22}$", r"$\beta_1$", r"$\beta_2$"]

fig, ax = plt.subplots(1, 1, figsize=(12,8))
_ = ax.boxplot(estimations_nonnoised)#, showmeans=True)
_ = ax.scatter(range(1, 9), single_theta[:-1], marker="*", s=100)
ax.set_ylim((0,3))
ax.set_xticklabels(parameters_full)
print("Proportion of null:", np.mean(estimations_nonnoised < 2e-16, axis=0))
_ = ax.set_title("Unnoised model")
```
